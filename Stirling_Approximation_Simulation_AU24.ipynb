```python
## Libraries

import math
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

from scipy.optimize import curve_fit
```

```python
def stirling(n):
    # Return estimation of ln n!
    return n*np.log(n) - n
def ln_factorial(n):
    # Return precise calculation of log factorial
    return np.cumsum(np.log(n_range))
def func(x, c, d):
    return c*np.log(d*x)

n_range = np.arange(1,10001)
s = stirling(n_range)
f = ln_factorial(n_range)

guess = (1.0,1.0)
fit, covar = curve_fit(func, n_range, f-s, guess)


fig, ax = plt.subplots(figsize=(4.5,7), dpi=300, nrows=2)
m = 60
n0, s0, f0 = n_range[:m], s[:m], f[:m]
max_pt = np.max(np.concatenate([s0,f0]))
#ax.plot([0,max_pt], [0,max_pt])
ax[0].plot(n0, f0, label='Exact')
ax[0].plot(n0, s0, label='Stirling')
ax[0].set_xlabel('N')
ax[0].set_ylabel('ln(N!)')
ax[0].legend()



ax[1].plot(n_range, f-s, label='Observed', linewidth=3)
ax[1].set_xlabel('N')
ax[1].set_ylabel('Exact - Stirling')


ax[1].plot(n_range, func(n_range,*fit), linestyle='dotted', label='Fit', linewidth=3)
ax[1].legend()
fig.tight_layout()
print(fit)
```
